**Importing libraries**

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from google.colab.patches import cv2_imshow
```

**Importing Sample Image**

```python
img = cv2.imread("board.jpg")
cv2_imshow(img)
```

**Converting image into text**

```python
txt = pytesseract.image_to_string(img)
print(txt)
```

**Creating boxes for the detected text and putting its detected text**

```python
box = pytesseract.image_to_boxes(img)
print(box)
```

```python
height,width,_ = img.shape
for detect in box.splitlines():
    detect = detect.split(" ")
    x,y,w,h = int(detect[1]),int(detect[2]),int(detect[3]),int(detect[4])
    cv2.rectangle(img,(x,height-y),(w,height-h),(0,0,255),3 ) ##using blue color for the box
    cv2.putText(img,detect[0],(x,height-y+25),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2) ##using red color for the detected text

plt.figure(figsize=(30,10))
plt.imshow(img)
```
